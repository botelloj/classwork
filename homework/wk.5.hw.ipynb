```python
from typing import List 
import pandas as pd
```

```python
names = ["Nate", "John", "Jane", "Mandy"]
age = [39, 23, 31, 25]
sex = ["m", "m", "f", "f"]
email = ["nate@gmail.com", "john@gmail.com", "jane@gmail.com", "mandy@gmail.com"]
purchases = [34, 12, 42, 8]

people = {
    "name" : names, 
    "age" : age, 
    "sex" : sex, 
    "email" : email,
    "purchases" : purchases 
}

people_df = pd.DataFrame(people)
people_df
```

```python
people_df = people_df.rename(columns={"sex" : "gender"})
people_df
```

```python
people_df["active"] = [True, True, True, True]
people_df
```
