```python
# Conditionals: == (equals), < (less than), <= (less than/equal to), > (greater than), >= (greater than/equal to),!= (not equal)

a:int = 10
b:int = 12

print(f"a={a}, b={b}")
print(f"a==b: {a==b}")
print(f"a>b: {a>b}")
print(f"a<=b: {a<=b}")
```

```python
from typing import List, Dict, Tuple, Set 
# type hinting: str, int, float, bool
# in, or, and, is

c: List[int] = [1,3,10,6,7] # for multiples list[str | int]

# in operator
print(f"is {a} in {c}: {a in c}")

# and operator
print(a==b and a in c) # cannot do this: a > b and 5 

# or operator 
print(a==b or a in c)

# is operator 
print(type(a) is int)

# print(type(a) == int)

print(a != b)

# not 
print(not True)
print (not False)
print(f"not false: {not False}")

print(f"not a < b: {not a<b}")
```

```python
# truthy / falsy

# falsy = 0, 0.0, "", None, [], {}, ()

print(bool(a))
print(bool(b))
print(bool(c))
print(bool(0))
print(bool([]))
```

```python
# try except 
try: 
    num: str = input("enetr a number b/w 1 & 10")
    if num.isnumeric():
        num = int(num)
        if num < 1 or num > 10:
            print("yey")
        else: 
            raise Exception("follow directions. i said between 1 and 10")
    else: 
        raise Exception("choose a number!")
except Exception as e:
    print(f"there was an error: {e}")
```
