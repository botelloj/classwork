```python
for idx, id, n, p, y, lt, lg, _ in df.itertuples():
    print(p)
```

```python
# rename Denmark with itertuples 
for item in df.itertuples():
    if item.place == "Denmark":
        df.at[item.Index, "place"] = "Den."
        
df["place"].unique()
```

```python
# df.loc[] to rename only one variable name 

df.loc[df["place"] == "England", "place"] = "Eng."
df["place"].unique()
```

```python
# hash = {
    # "Geneva" : "Gen.",
    # "Grisons" : "Gri",
# }

hash = {
    place: f"{place[:3]}." for place in df["place"].unique()
    }

df["place"] = df["place"].replace(hash)
df["place"].unique()
```

```python
df.loc[(df["place"] == "Eng.") & (df["year"] >= 1550) & (df["year"] <= 1555)].count()
```

```python
df.loc[~(df["place"] == "Eng.") & (df["year"] >= 1550) & (df["year"] <= 1555)].count()
```

```python
imp_emb = ["Ven.", "Eng.", "Ott."]

imp_num = df.loc[(df["place"].isin(imp_emb)) & (df["year"] >= 1550) & (df["year"] <= 1555)].count()
non_imp_num = df.loc[~(df["place"].isin(imp_emb)) & (df["year"] >= 1550) & (df["year"] <= 1555)].count()
all_num = df.loc[(df["year"] >= 1550) & (df["year"] <= 1555)].count()

print(f"all ambs: {all_num}")
print(f"% to imp: {imp_num / all_num}")
print(f"% to non_imp: {non_imp_num / all_num}")

```
