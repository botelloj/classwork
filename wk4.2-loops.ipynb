```python
# while loop 
count = 0
while True: # while this condition is true 
    # run this code block over and over 
    if count > 10: 
        break 
    print(count) 
    count += 1 # count = count + 1 
```

```python
# for loop itersates over collections 

from typing import List 
collection: List[int] = [1, 2, 3, 4, 5]

for num in collection: 
    print(num)
    
# to get index and item, use enumerate 
for item in enumerate(collection):
    print(f"{item[1]} is at index {item[0]}")

# deconstruct the variable because enumerate returns (index, seq[index])
for idx, num in enumerate(collection): 
    print(f"{num} is at index {idx}")

```

```python
sentence: str = "This is a Python course"
words: List[str] = sentence.split(" ") # split returns a list
used_vowels: List[str] = []
vowels_options: List[str] = ["a", "e", "i", "o", "u", "y"]
for word in words:
    for letter in word:
        if letter in vowels_options:
            used_vowels.append(letter)
            
print(used_vowels)
```

```python
# functions take an input (arguements)
# process it, and return an output (return)
def shorten_names(names_list: List[str], chars: int) -> List[str]:
    """
    Shortens the strings inside a list to three characters.
    
    Args: 
        names_list (List[str]): the list of strings you want
        shortened to three characters.
        chars (int): the number of characters to shorten 
        each string in the list to.
        
    Return:
        List[str]:a new list of shortened strings.
    """
    # cannot read information from outside the function
    names_list =  [f"{name[:chars]}." for name in names_list]
    return names_list

shorten_names(countries, 2)
shorten_names(countries, chars=2)
```
